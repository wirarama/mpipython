```python
from bs4 import BeautifulSoup
import requests as req
```

```python
def caridetik(url,page,konten):
    resp = req.get(url)
    soup = BeautifulSoup(resp.text,'lxml')
    for a in soup.find_all('a',href=True):
        if("detik.com" in a['href']):
            suburl = a['href'].split('/')
            if len(suburl)>4:
                if suburl[3]=='search' and '&page=' in a['href']:
                    if(a['href'] not in page):
                        page.append(a['href'])
                elif len(suburl)>5:
                    if((a['href'],suburl[3]) not in konten):
                        konten.append((a['href'],suburl[3]))
    return page,konten
```

```python
page = []
konten = []
page,konten = caridetik("https://www.detik.com/search/searchall?query=corona+indonesia",page,konten)

print(page)
print(konten)
```

```python
for pp in page:
    print(pp)
    page,konten = caridetik(pp,page,konten)
```

```python
print(len(page))
print(len(konten))
```

```python
pagefile = open('detik/page.txt','w')
for x in page:
    pagefile.write(x.replace(' ','+')+"\n")
pagefile.close()
```

```python
kontenfile = open('detik/konten.txt','w')
for x in konten:
    kontenfile.write(x[0].replace(' ','+')+","+x[1]+"\n")
kontenfile.close()
```

```python
for x in konten:
    f = open('detik/kategori/'+x[1]+'.txt','a')
    f.write(x[0].replace(' ','+')+"\n")
    f.close()
```

```python
def detailberitadetik(url):
    resp = req.get(url)
    soup = BeautifulSoup(resp.text,'lxml')
    out = ""
    for judul in soup.find_all('h1',{"class":"detail__title"}):
        out += judul.text
    for isi in soup.find_all('div',{"class":"detail__body-text"}):
        out += isi.text
    return out
```
